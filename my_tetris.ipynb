```python
import py5 

class BigBoard:
    def __init__(self):
        self.w = 20 #cell width
        self.columns = 12 #num cols
        self.rows = 20 #num rows


    def create_2d_list(self, cols, rows): #creates a 2d list filled with 0's
        lst = []
        for i in range(cols):
            lst.append([0] * rows)
        return lst
    
    def setup(self):
        self.board = self.create_2d_list(self.columns, self.rows)
        for i in range(self.columns):
            for j in range(self.rows):
                self.board[i][j] = 0

    def draw(self):
        py5.background(255) #white background
        for i in range(self.columns):
            for j in range(self.rows):
                py5.stroke(0) #black border
                py5.stroke_weight(1)

                if self.board[i][j] == 0:
                    py5.fill(255) #white fill
                elif self.board[i][j] == 1: #t
                    py5.fill(106, 13, 173)
                elif self.board[i][j] == 2: #o
                    py5.fill(255, 251, 0)
                else:# self.board[j][i] == 3: #i
                    py5.fill(173, 216, 230)
                py5.square(i*self.w, j*self.w, self.w) #square with width w at position depending on "board"
    
    def add_shape(self, shape):
        row = int(shape.row)
        if shape.type == 't':
            self.board[shape.col][row] = 1
        elif shape.type == 'o':
            self.board[shape.col][shape.row] = 2
        else:
            self.board[int(shape.col)][row] = 3

class Shapes:
    def __init__(self, type, orientation, row, col, bigboard):
        self.w = 20
        self.type = type #o, i, s, z, j, l, t
        self.o = orientation #orientation, 0 up, 1 right, 2 down, 3 left
        self.row = row
        self.col = col
        self.bigboard = bigboard  
    
    def draw(self):
        row = self.row
        col = self.col
        if self.type == 't':
            self.draw_t(row, col)
        elif self.type == 'o':
            self.draw_o(row, col)
        elif self.type == 'i':
            self.draw_i(row, col)
        elif self.type == 's':
            self.draw_s(row, col)
        elif self.type == 'z':
            self.draw_z(row, col)
        elif self.type == 'j':
            self.draw_j(row, col)
        elif self.type == 'l':
            self.draw_l(row, col)
    

    def draw_t(self, row, col):
        if self.row > 18:
            self.row = 18
        if self.col < 1:
            self.col = 1
        if self.col > 11:
            self.col = 11
        py5.stroke(0) #black border
        py5.stroke_weight(1)
        py5.fill(106, 13, 173)
        if self.o == 0:
            py5.square(col*self.w, row * self.w, self.w)
            py5.square(col*self.w, (row+1)*self.w, self.w)
            py5.square((col-1)*self.w, (row+1)*self.w, self.w)
            py5.square((col+1)*self.w, (row+1)*self.w, self.w)

        elif self.o == 1:
            py5.square((col-1)*self.w, row*self.w, self.w)
            py5.square((col-1)*self.w, (row+1)*self.w, self.w)
            py5.square((col-1)*self.w, (row+2)*self.w, self.w)
            py5.square(col*self.w, (row+1)*self.w, self.w)

        elif self.o == 2:
            py5.square(col*self.w, row*self.w, self.w)
            py5.square((col-1)*self.w, row*self.w, self.w)
            py5.square((col+1)*self.w, row*self.w, self.w)
            py5.square(col*self.w, (row+1)*self.w, self.w)

        else: 
            py5.square(col*self.w, row*self.w, self.w)
            py5.square(col*self.w, (row+1)*self.w, self.w)
            py5.square(col*self.w, (row+2)*self.w, self.w)
            py5.square((col-1)*self.w, (row+1)*self.w, self.w)
    
    def draw_o(self, row, col):
        if self.row > 18:
            self.row = 18
        if self.col < 1:
            self.col = 1
        if self.col > 11:
            self.col = 11
        py5.stroke(0) #black border
        py5.stroke_weight(1)
        py5.fill(255, 251, 0)
        py5.square(self.w * col, row*self.w, self.w)
        py5.square(self.w * (col - 1), row*self.w, self.w)
        py5.square(self.w * col, (row+1)*self.w, self.w)
        py5.square(self.w * (col - 1), (row+1)*self.w, self.w)
    
    def draw_i(self, row, col):
        if self.row > 18:
            self.row = 18
        if self.col < 0:
            self.col = 0
        if self.col > 11:
            self.col = 11
        py5.stroke(0) #black border
        py5.stroke_weight(1)
        py5.fill(173, 216, 230)
        if self.o == 0 or self.o == 2:
            py5.square(self.w * (col ), row * self.w, self.w)
            py5.square(self.w * (col ), (row + 1)*self.w, self.w)
            py5.square(self.w * (col ), (row + 2)*self.w, self.w)
            py5.square(self.w * (col ), (row + 3)*self.w, self.w)

        elif self.o == 1 or self.o == 3:
            py5.square(self.w * col, (row + 1) * self.w, self.w)
            py5.square(self.w * (col - 1), (row + 1) * self.w, self.w)
            py5.square(self.w * (col - 2), (row + 1)* self.w, self.w)
            py5.square(self.w * (col + 1), (row + 1) * self.w, self.w)
    
    def draw_s(self, row, col):
        if self.row > 18:
            self.row = 18
        if self.col < 1:
            self.col = 1
        if self.col > 11:
            self.col = 11
        py5.stroke(0) #black border
        py5.stroke_weight(1)
        py5.fill(255, 200, 200)
        if self.o == 0 or self.o == 2:
            py5.square(self.w * col, row * self.w, self.w)
            py5.square(self.w * (col+1), row * self.w, self.w)
            py5.square(self.w * col, (row+1)*self.w, self.w)
            py5.square(self.w * (col - 1), (row+1)*self.w, self.w)

        elif self.o == 1 or self.o == 3:
            py5.square(self.w * (col - 1), row * self.w, self.w)
            py5.square(self.w * (col - 1), (row + 1)*self.w, self.w)
            py5.square(self.w * col, (row + 1)*self.w, self.w)
            py5.square(self.w * col, (row + 2)*self.w, self.w)

    def draw_z(self, row, col):
        if self.row > 18:
            self.row = 18
        if self.col < 1:
            self.col = 1
        if self.col > 11:
            self.col = 11
        py5.stroke(0) #black border
        py5.stroke_weight(1)
        py5.fill(144, 238, 144)
        if self.o == 0 or self.o == 2:
            py5.square(self.w * col, row * self.w, self.w)
            py5.square(self.w * (col - 1), row * self.w, self.w)
            py5.square(self.w * col, (row+1)*self.w, self.w)
            py5.square(self.w * (col + 1), (row+1)*self.w, self.w)

        elif self.o == 1 or self.o == 3:
            py5.square(self.w * col, row * self.w, self.w)
            py5.square(self.w * col, (row+1)*self.w, self.w)
            py5.square(self.w * (col - 1), (row+1)*self.w, self.w)
            py5.square(self.w * (col - 1), (row+2)*self.w, self.w) 
    
    def draw_j(self, row, col):
        if self.row > 18:
            self.row = 18
        if self.col < 1:
            self.col = 1
        if self.col > 11:
            self.col = 11
        py5.stroke(0) #black border
        py5.stroke_weight(1)
        py5.fill(255, 102, 178)
        if self.o == 0:
            py5.square(self.w * col, (row -1) * self.w, self.w)
            py5.square(self.w * col, (row)*self.w, self.w)
            py5.square(self.w * col, (row + 1)*self.w, self.w)
            py5.square(self.w * (col - 1), (row + 1)*self.w, self.w)

        elif self.o == 1:
            py5.square(self.w * (col -1), row * self.w, self.w)
            py5.square(self.w * (col -1), (row + 1)*self.w, self.w)
            py5.square(self.w * col, (row + 1)*self.w, self.w)
            py5.square(self.w * (col +1), (row + 1)*self.w, self.w)
        
        elif self.o == 2:
            py5.square(self.w * col, (row - 1) * self.w, self.w)
            py5.square(self.w * (col -1), (row - 1) * self.w, self.w)
            py5.square(self.w * (col -1), (row)*self.w, self.w)
            py5.square(self.w * (col -1), (row + 1)*self.w, self.w)
        
        elif self.o == 3:
            py5.square(self.w * (col -1), row * self.w, self.w)
            py5.square(self.w * col, row * self.w, self.w)
            py5.square(self.w * (col +1), row * self.w, self.w)
            py5.square(self.w * (col +1), (row + 1)*self.w, self.w)
    
    def draw_l(self, row, col):
        if self.row == -99:
            x = 0 # do nothing
        else:
            if self.row > 18:
                bigboard.add_shape(self)
                self.row = -99 #18
            if self.col < 1:
                self.col = 1
            if self.col > 11:
                self.col = 11
            py5.stroke(0) #black border
            py5.stroke_weight(1)
            py5.fill(255, 178, 102)
            if self.o == 0:
                py5.square(self.w * (col -1), (row-1) * self.w, self.w)
                py5.square(self.w * (col -1), (row)*self.w, self.w)
                py5.square(self.w * (col -1), (row+1)*self.w, self.w)
                py5.square(self.w * col, (row+1)*self.w, self.w)

            elif self.o == 1:
                py5.square(self.w * (col -1), row * self.w, self.w)
                py5.square(self.w * (col), row * self.w, self.w)
                py5.square(self.w * (col +1), row * self.w, self.w)
                py5.square(self.w * (col -1), (row+1)*self.w, self.w)
            
            elif self.o == 2:
                py5.square(self.w * (col -1), (row -1) * self.w, self.w)
                py5.square(self.w * col, (row)*self.w, self.w)
                py5.square(self.w * col, (row+1)*self.w, self.w)
                py5.square(self.w * col, (row -1) * self.w, self.w)
            
            elif self.o == 3:
                py5.square(self.w * (col +1), row * self.w, self.w)
                py5.square(self.w * (col +1), (row+1)*self.w, self.w)
                py5.square(self.w * col, (row+1)*self.w, self.w)
                py5.square(self.w * (col -1), (row+1)*self.w, self.w)
            

class Game:
    def __init__(self, bigboard, shapes):
        self.bigboard = bigboard
        self.shapes = shapes
        self.score = 0
        self.level = 1
        self.lines_cleared = 0
    
    def draw(self):
        self.shapes.draw()
        self.shapes.row += 1


bigboard = BigBoard()
shapes = Shapes("l", 0, 0, bigboard.columns * .5, bigboard) #o, i, s, z, j, l, t
game = Game(bigboard, shapes)

def setup():
    py5.size(600, 600)
    py5.background(255)
    py5.frame_rate(1)
    bigboard.setup()

    

def draw():
    bigboard.draw()
    game.draw()


def mouse_clicked():
    x = 1
    #shapes.o = (shapes.o + 1) % 4

def key_pressed():
    if py5.key == py5.CODED:
        if py5.key_code == py5.UP:
            shapes.o = (shapes.o + 1) % 4
        if py5.key_code == py5.DOWN:
            shapes.row += 1
        if py5.key_code == py5.LEFT:
            shapes.col -= 1
        if py5.key_code == py5.RIGHT:
            shapes.col += 1
    if py5.key == ' ':
        if shapes.row > 0:
            bigboard.add_shape(shapes)
        #new shape


py5.run_sketch()
```
