```python
import py5

class TheBoard:
    def __init__(self):
        self.w = 20 #cell width
        self.columns = 12 #num cols
        self.rows = 20 #num rows


    def create_2d_list(self, cols, rows): #creates a 2d list filled with 0's
        lst = []
        for i in range(cols):
            lst.append([0] * rows)
        return lst
    
    def draw(self):
        py5.background(255) #white background
        for i in range(self.columns):
            for j in range(self.rows):
                py5.stroke(0) #black border
                py5.stroke_weight(1)

                if self.board[i][j] == 0:
                    py5.fill(255) #white fill
                elif self.board[i][j] == 9:
                    py5.fill(255,  0, 0)
                else:
                    py5.fill(173, 216, 230) #other fill
                py5.square(i*self.w, j*self.w, self.w) #square with width w at position depending on "board"

    def setup(self):
        self.board = self.create_2d_list(self.columns, self.rows)
        for i in range(self.columns):
            for j in range(self.rows):
                self.board[i][j] = 0      

    def draw_shape(self, shape, type, o): 
        if shape.row>= 19:
            shape.row = 19
            shape.can_move = False
        if shape.row < 0:
            shape.row = 0
        if shape.col < 0:                
            shape.col = 0
        if shape.col > 11:
            shape.col = 11
        
        check_below = self.check_below(shape.row, shape.col)
        if check_below == False:
            shape.can_move = False

        self.board[shape.col][shape.row] = 1
            #self.board[col-1][row] = 1
            #self.board[col][row+1] = 1
            #self.board[col-1][row+1] = 1
    
    def update_board(self, row, col):
        self.board[col][row] = 0

    def check_below(self, row, col): #if the spot directly below is taken return False
        result = True
        if row < 19:
            if self.board[col][row + 1] == 1: #if spot below is filled
                result = False
            else:
                result = True
        else:
            result = False
        return result

    def lowest_free_space(self, row, col):
        for i in range(row, 19): #from current row to bottom
            if self.board[col][i + 1] == 1: #if spot below is filled
                return i
        return 19

    def full_col(self): #will determine if all cells in a column are filled
        for i in range(self.columns):
            full = True
            for j in range(self.rows):
                if self.board[i][j] == 0:
                    full = False
            if full:
                return i
        return -1
    
    def full_row(self): #will determine if all cells in a row are filled
        for j in range(self.rows):
            full = True
            for i in range(self.columns):
                if self.board[i][j] == 0:
                    full = False
            if full:
                return j
        return -1
        

class Shape:
    def __init__(self, type, o, row, col):
        self.type = type
        self.o = o
        self.row = row
        self.col = col
        self.can_move = True
    
    def update(self, o, row, col):
        self.o = o
        self.row = row
        self.col = col

theboard = TheBoard()
shapes = []
shape = Shape("o", 0, 0, 6)
shapes.append(shape)

def count_can_move(shapes):
    count = 0
    for shape in shapes:
        if shape.can_move:
            count += 1
    return count

def setup():
    py5.size(600, 600)
    py5.background(255)
    theboard.setup()   

def draw():
    for shape in shapes:
        theboard.draw_shape(shape, shape.type, shape.o)
    theboard.draw()

    can_move_total = count_can_move(shapes)
    if can_move_total == 0:
        if theboard.full_col() == -1:
            #before we add a new shape, check for full rows
            full_row = theboard.full_row()
            while full_row != -1:
                for i in range(theboard.columns):
                    theboard.board[i].pop(full_row) #remove that row from each column
                    theboard.board[i].insert(0, 0) #add a new empty row at the top of each column
                full_row = theboard.full_row()
#!!! it's not clearing the board properly
            shape = Shape("o", 0, 0, 6)
            shapes.append(shape)
        else:
            print("Game Over")
            py5.no_loop()

def key_pressed():
    for shape in shapes:
        if shape.can_move:
            if py5.key == py5.CODED:
                    
                    if py5.key_code == py5.UP:
                        if theboard.board[shape.col][shape.row -1] ==0:
                            theboard.update_board(int(shape.row), int(shape.col))
                            shape.update(shape.o, shape.row - 1, shape.col)
                    if py5.key_code == py5.DOWN:
                        if theboard.board[shape.col][shape.row +1] ==0:
                            theboard.update_board(int(shape.row), int(shape.col))
                            shape.update(shape.o, shape.row + 1, shape.col)
                    if py5.key_code == py5.LEFT:
                        if theboard.board[shape.col -1][shape.row] ==0:
                            theboard.update_board(int(shape.row), int(shape.col))
                            shape.update(shape.o, shape.row, shape.col -1)
                    if py5.key_code == py5.RIGHT:
                        if theboard.board[shape.col +1][shape.row] ==0:
                            theboard.update_board(int(shape.row), int(shape.col))
                            shape.update(shape.o, shape.row, shape.col +1)
        
                    theboard.draw_shape(shape, shape.type, shape.o)
            if py5.key == ' ':
                lowest_space_free = theboard.lowest_free_space(shape.row, shape.col)
                theboard.update_board(int(shape.row), int(shape.col))
                shape.update(shape.o, lowest_space_free, shape.col)
                theboard.draw_shape(shape, shape.type, shape.o)

py5.run_sketch()
```
